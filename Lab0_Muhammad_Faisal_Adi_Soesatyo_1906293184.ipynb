Import Library

```python
from skimage import color, io
from skimage.transform import rescale
import matplotlib.pyplot as plt
```

Read Image

```python
image_1 = io.imread('fasilkom-building.jpg')
transform_i1 = color.rgb2gray(image_1)
```

Show Image

```python
plt.imshow(transform_i1, cmap='gray', vmin=0, vmax=1)
plt.show()
```

Image Dimension Checking

```python
dimension = transform_i1.shape
dimension
```

Saving Image

```python
save_image_1 = io.imsave('fasilkom-building-gray.jpg', transform_i1)
save_image_1
```

Resolution

```python
image_3 = rescale(transform_i1, .5, anti_aliasing=True)
plt.imshow(image_3, cmap='gray')
```

```python
image_4 = rescale(transform_i1, .07, anti_aliasing=True)
plt.imshow(image_4, cmap='gray')
```

Image Conversion

```python
original = plt.imshow(image_1)
```

```python
rgb2hsv = color.hsv2rgb(image_1)
plt.imshow(rgb2hsv)
```

```python
rgb2gray = color.rgb2gray(image_1)
plt.imshow(rgb2gray, cmap='gray')
```
