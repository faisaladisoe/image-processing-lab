## Import package

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, io, util, exposure, filters, morphology
```

# Soal 1

## Load and Show Image

```python
mask_image = io.imread('./images/mask.jpg')
plt.imshow(mask_image, cmap = 'gray', vmin = 0, vmax = 255)
plt.show()
```

## Show Histogram

```python
plt.figure(figsize=(15, 10))

mask_image_flat = util.img_as_ubyte(mask_image).flatten()
plt.subplot(2,1,1)
plt.hist(mask_image_flat, 255, range=(0,255))
plt.title('Original', fontsize=20)
```

## Histogram Equalization

```python
mask_image_eq = exposure.equalize_hist(mask_image)

plt.figure(figsize=(15, 10))

plt.subplot(1,2,1); plt.imshow(mask_image)
plt.title('Original', fontsize=20); plt.axis("off")

plt.subplot(1,2,2); plt.imshow(mask_image_eq)
plt.title('Histogram Equalization', fontsize=20);
plt.axis("off")

plt.show()
```

### Histogram

```python
plt.figure(figsize=(15, 10))

mask_image_flat = util.img_as_ubyte(mask_image).flatten()
plt.subplot(2,1,1)
plt.hist(mask_image_flat, 256, range=(0,256))
plt.title('Original', fontsize=20)

mask_image_eq_flat = util.img_as_ubyte(mask_image_eq).flatten()
plt.subplot(2,1,2)
plt.hist(mask_image_eq_flat, 256, range=(0,256))
plt.title('Histogram Equalization', fontsize=20)

plt.subplots_adjust(hspace=0.3)
plt.show()
```

## Contrast Stretching

```python
plt.figure(figsize=(15, 10))

mn = min(mask_image.flatten())
mx = max(mask_image.flatten())
b = int(np.floor(255 / (mx - mn)))
mask_image_cs = (mask_image - mn) * b

plt.subplot(1,2,1); plt.imshow(mask_image)
plt.title('Original', fontsize=20); plt.axis("off")

plt.subplot(1,2,2); plt.imshow(mask_image_cs)
plt.title('Contrast Stretching', fontsize=20);

plt.axis("off")
plt.show()
```

### Histogram

```python
plt.figure(figsize=(15, 10))

mask_image_flat = util.img_as_ubyte(mask_image).flatten()
plt.subplot(2,1,1)
plt.hist(mask_image_flat, 256, range=(0,256))
plt.title('Original', fontsize=20)

mask_image_cs_flat = util.img_as_ubyte(mask_image_cs).flatten()
plt.subplot(2,1,2)
plt.hist(mask_image_cs_flat, 256, range=(0,256))
plt.title('Contrast Stretching', fontsize=20)

plt.subplots_adjust(hspace=0.3)
plt.show()
```

## Difference between histogram equalization and contrast Stretching

### Histogram Equalization
1.   Berdasarkan hasil histogram yang saya hasilkan, jarak antar bar cenderung lebih renggang serta terdapat beberapa bar yang berkelompok.
2.   Dari pemaparan point 1, dapat dimaknai sebagai hasil histogram equalization warnanya lebih kontras serta pada beberapa bagian terdapat warna yang lebih padat ketimbang bagian lainnya.
3.   Dilihat dari pola penyebaran bar pada histogram, dapat disimpulkan bahwa teknik histogram equalization bekerja dengan cara mencari bar yang padat dan sering muncul, kemudian bar-bar tersebut disebar ke range yang lebih sparse.

### Contrast Stretching
1.   Berdasarkan hasil histogram yang saya hasilkan, posisi antar bar cenderung memiliki jarak yang sama satu dengan lainnya. Serta tidak ditemui bagian bar yang berkelompok.
2.   Dari pemaparan point 1, dapat dimaknai sebagai hasil contrast stretching lebih vary karena persebaran cukup merata dari 0-255; Serta tidak ada bagian yang memiliki warna lebih padat & kontras ketimbang bagian lainnya.
3.   Dilihat dari pola penyebaran bar pada histogram, dapat disimpulkan bahwa teknik contrast stretching tidak melihat area mana yang lebih padat maupun bar pada value mana yang lebih sering muncul, objective utamanya adalah menyebarkan bar-bar tersebut ke range yang lebih sparse serta menjaga jarak antar bar agar tidak terlalu jauh maupun dekat.

# Soal 3

## Load and Show Image

```python
cat_image = util.img_as_ubyte(color.rgb2gray(io.imread('./images/cat.jpg')))
plt.imshow(cat_image, cmap = 'gray', vmin = 0, vmax = 255)
plt.show()
```

## Mean Filter

```python
fi_mean = filters.rank.mean(cat_image,selem=morphology.square(3))

plt.figure(figsize=(15, 10))

plt.subplot(1,2,1)
plt.imshow(cat_image,cmap='gray',vmin=0,vmax=255)
plt.title('Original', fontsize=20); plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(fi_mean,cmap='gray',vmin=0,vmax=255)
plt.title('Mean Filtered', fontsize=20); plt.axis("off")
plt.show()

plt.show()
```

## Median Filter

```python
fi_median = filters.rank.median(cat_image,selem=morphology.square(3))

plt.figure(figsize=(15, 10))

plt.subplot(1,2,1)
plt.imshow(cat_image,cmap='gray',vmin=0,vmax=255)
plt.title('Original', fontsize=20); plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(fi_median,cmap='gray',vmin=0,vmax=255)
plt.title('Median Filtered', fontsize=20); plt.axis("off")
plt.show()

plt.show()
```

## Difference between mean filter and median filter

### Mean Filter
1.   Hasil dari mean filter cukup mengurangi jumlah noise, dalam hal ini black dots (salt & pepper). Namun apabila melihat hasil filteringnya, secara sekilas hampir tidak ada bedanya antara hasil filtering dengan original image. Karena filtering yang dilakukan kurang sempurna 
2.   Pada mean filter, gambar hasil filter terlihat sedikit lebih gelap pada bagian kucingnya. Entah ini ilusi optik atau bukan, namun secara sekilas terlihat demikian 

### Median Filter
1.   Berbeda dengan mean filter, hasil filtering dari median filter jauh lebih sempurna dalam menghilangkan noise berupa black dots (salt & pepper). Serta tetap mempertahankan edge dari objects yang ada di dalam gambar.
2.   Sedangkan pada median filter, gambar yang dihasilkan lebih cerah ketimbang mean filter dan original imagenya. Entah karena efek black dots yang sudah hilang atau bukan, namun hasilnya lebih tajam, cerah dan halus.

Filter yang lebih bagus menurut saya adalah median filter. Berdasarkan cara kerja filtering yang saya pahami, median filter bekerja dengan cara mengurutkan ascending nilai yang ada pada area kernel 3x3, kemudian mencari mediannya. Sedangkan mean, mencari rata-rata dari area kernel 3x3. Drawback dari mean filtering adalah apabila terdapat satu atau dua nilai yang sangat berbeda jauhnya ketimbang tetangga lainnya, akan mempengaruhi hasil akhir dari filteringnya. Berbeda dengan median filter, tak pengaruh apakah ada nilai yang 'jomplang' bedanya, yang terpenting tetap mencari median valuenya. Selain itu, pada median filter tidak generate pixel value yang baru, median filter hanya menggunakan pixel yang ada. Berbeda dengan mean filter yang generate new pixel value hasil operasi averaging pixel value pada area kernelnya. Hal ini yang menyebabkan hasil dari median filter lebih tajam & halus ketimbang mean filter.
